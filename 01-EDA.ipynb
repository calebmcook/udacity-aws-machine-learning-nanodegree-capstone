```python
import pandas as pd
import os
import glob
```

## Data import

```python
data_path = os.path.join(os.getcwd(), 'use_this_data', '*')
```

```python
files = glob.glob(data_path)
```

```python
files
```

```python
wl = pd.read_parquet(files[0], columns=['Security Code Clean', 'file_date'])
rat = pd.read_parquet(files[1], columns=['Security Code Clean', 'file_date', 'P/E ratio', 'P/B ratio'])
sq = pd.read_parquet(files[2],  columns=['Security Code Clean', 'file_date', 'Trade Volume', 'Closing Price', 'Change'])
ss = pd.read_parquet(files[3], columns=['Security Code Clean', 'file_date', 'Trading Volume', 'Trading Value'])
```

```python
sq
```

```python
wl
```

```python
rat
```

```python
ss.rename(columns={'Trading Volume':'ss_trading_vol', 'Trading Value':'ss_trading_value'}, inplace=True)
```

```python
ss
```

```python
all1 = pd.merge(left=sq, right=ss, how='left', on=['Security Code Clean', 'file_date'], suffixes=('_sq','_ss'))
```

```python
all2 = pd.merge(left=all1, right=rat, how='left', on=['Security Code Clean', 'file_date'], suffixes=('_sq_ss', '_rat'))
```

```python
all = pd.merge(left=all2, right=wl, how='left', on=['Security Code Clean', 'file_date'], suffixes=('_sq_ss_rat', '_wl'))
```

```python
all
```

```python
all.set_index('file_date', inplace=True)
```
